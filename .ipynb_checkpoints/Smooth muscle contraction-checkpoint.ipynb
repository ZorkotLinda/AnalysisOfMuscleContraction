```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
%matplotlib inline
```

```python
contrac = np.loadtxt('contraction.csv', delimiter=',')
```

## Info about experiment
Duration of the experiment **62 minutes** <br>
Stabilization **0-30 minutes** <br>
KCL application **31-37 minutes** <br>
Carbacholine application **56-62 minutes**

Let's determine time step Δt:

```python
Δt = 62 / len(contrac)
print('time step Δt = %s' % Δt)
```

Construct our time vector:

```python
time = np.arange(0, 62, Δt)
```

```python
fig, axs = plt.subplots(figsize=(20, 4), dpi=130)
axs.plot(time, contrac)
axs.axvline(x=30, color='red', label='KCL')
axs.axvline(x=37, color='red',)
axs.axvline(x=56, color='green', label='Carbacholine')
axs.axvline(x=62, color='green')
axs.set(xlabel='time, mins')
axs.legend(title='Apllication')
axs.set_title('Contraction of myometrium over time')
```

```python
from scipy.signal import find_peaks, peak_widths
from scipy.integrate import simps
```

```python
peaks, _ = find_peaks(Imodpoly_contrac, height=0.25, distance=300)
mins = argrelmin(Imodpoly_contrac)
widths, width_heights, left, right = peak_widths(Imodpoly_contrac, peaks, rel_height=0.5)
widths_full, width_heights_full, left_full, right_full = peak_widths(Imodpoly_contrac, peaks, rel_height=0.95)
#print(widths)
```

```python
fig, axs = plt.subplots(figsize=(20, 4), dpi=130)
axs.plot(time, Imodpoly_contrac)
axs.plot(time[peaks], Imodpoly_contrac[peaks], 'x')
axs.hlines(width_heights, left*Δt, right*Δt, color="red")
axs.plot(left_full*Δt, width_heights_full, 'x', color='red')
axs.plot(right_full*Δt, width_heights_full, 'x', color='red')
axs.set(xlabel='time, s')
#axs.set_xlim([0, 10])
#axs.set_ylim([0, .8])
axs.set_title('Contraction of myometrium over time')
```

```python
half_widths = widths * Δt
left_bounds = left_full * Δt
right_bounds = right_full * Δt
```

Calculating area under the peaks

Calculating amplitude

```python
amps = []
areas = []
for idx, peak in enumerate(peaks):
    neibs = np.diff( Imodpoly_contrac[peak-300:peak+300] ) 
    base = 0 
    amp = Imodpoly_contrac[peak] - base
    amps.append(amp)
    area = simps(Imodpoly_contrac[np.where( (time > left_bounds[idx]) & (time < right_bounds[idx]))],
                 time[np.where( (time > left_bounds[idx]) & (time < right_bounds[idx]))] * 60) 
    areas.append(area)
```

```python
print(areas)
```

```python
fig, axs = plt.subplots(figsize=(20, 4), dpi=130)
axs.plot(time, Imodpoly_contrac)
axs.plot(time[peaks], Imodpoly_contrac[peaks], 'x')
axs.axvline(x=30, color='red', label='KCL')
axs.axvline(x=56, color='green', label='Carbacholine')
axs.set(xlabel='time, s')
axs.set_xlim([0,30])
#axs.set_ylim([0, .8])
axs.legend(title='Application')
axs.set_title('Contraction of myometrium over time')
```

```python
amps_onset = amps[:np.argmax(time[peaks] > 30)]
amps_kcl = amps[np.argmax(time[peaks] > 30) : np.argmax(time[peaks] > 37)]
amps_carbocholine = amps[np.argmax(time[peaks] > 56):]
```

Plotting errorbars

```python
y = [np.mean(amps_onset), np.mean(amps_kcl), np.mean(amps_carbocholine)]
labels = ['Control', 'KCL', 'Carbacholine']
x = np.arange(len(labels))
error = [np.std(amps_onset), np.std(amps_kcl), np.std(amps_carbocholine)]
```

```python
fig, ax = plt.subplots(dpi=150);
ax.bar(x, y, yerr=error, align='center', alpha=0.8, ecolor='red', capsize=10);
ax.set(ylabel='Amplitude, g');
ax.set_xticks(x);
ax.set_title('Errorbars for myometrium contraction amplitude');
ax.set_xticklabels(labels);
```
